```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

 Create a Simple Dataset

```python
X = np.linspace(0, 10, 50).reshape(-1, 1)
X
```

```python
#what x represents?- a continuous growth variable
#why this range was chosen?- to show early rapid growth and later saturation
```

```python
noise = np.random.normal(0, 0.2, size=(50,1))
y = np.log(X + 1) + noise
```

 Decide the Model Shape

```python
#how many intermediate values want?- 3
#why not 1?  One hidden unit behaves almost linear, multiple units allow multiple bending regions.
#why not too many?- Too many increases instability and overfitting risk.
```

Initialize Parameters

```python
W1 = np.random.uniform(-1, 1, size=(1,3))
b1 = np.zeros((1,3))

W2 = np.random.uniform(-1, 1, size=(3,1))
b2 = np.zeros((1,1))
```

 Activation Function (ReLU)

```python
def activation(z):
    return np.maximum(0, z)

def activation_slope(z):
    return (z > 0).astype(float)
```

Training loop

```python
learning_rate = 0.01
epochs = 1000

for epoch in range(epochs):

    # Forward pass
    z1 = X @ W1 + b1
    h = activation(z1)
    y_hat = h @ W2 + b2

    # Loss
    error = y_hat - y
    loss = np.mean(error ** 2)

    # Backward pass
    dL_dy = 2 * error / len(X)

    dL_dW2 = h.T @ dL_dy
    dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)

    dL_dh = dL_dy @ W2.T

    dL_dz1 = dL_dh * activation_slope(z1)

    dL_dW1 = X.T @ dL_dz1
    dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)

    # Parameter update
    W1 -= learning_rate * dL_dW1
    b1 -= learning_rate * dL_db1
    W2 -= learning_rate * dL_dW2
    b2 -= learning_rate * dL_db2


    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

# Final Output
print("\nFinal Loss:", loss)
```
